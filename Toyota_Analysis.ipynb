```python
Weekly Assignment-1 Group 4
Arjun Aanand
Sumesh Kumar Jegadesh Kumar

Problem 2.1 
Answer- 
a: "Supervised Learning",
b: "Unsupervised Learning",
c: "Supervised Learning",
d: "Unsupervised Learning",
e: "Supervised Learning",
f: "Supervised Learning",
g: "Supervised Learning",
h: "Supervised Learning"

Problem 2.5
Overfitting and Zero Error in Training Data
Because of the phenomenon of overfitting, a model fitted to training data may not always produce zero errors. 
When a model learns the noise and random fluctuations in addition to the underlying patterns in the training data, this is known as overfitting. As a result, the model performs terribly on fresh, untested data but remarkably well on the training set. 
1.Overfitting Definition: An overfit model is one that identifies outliers and noise in the training data rather than focusing just on the broad trends. As a result, the model becomes overly intricate and tuned in too much to the training set.
2.Zero Error Concern: A model is likely too sophisticated and has learned the training data if it achieves zero error on the training set. Although this would appear ideal, it usually means that the model has not generalized very successfully.
3.Generalization: Applying the lessons learned from the training data to fresh data from the same distribution is the main objective of a prediction model. Overfitting a model will result in low bias but large variance, which will make it underperform on test data and in practical applications.
4.Bias-Variance Tradeoff: The bias-variance tradeoff and overfitting are connected. On unseen data, a highly complex model (with little bias) will perform poorly due to its huge variance. On the other hand, a more straightforward model may have greater bias but less variance, which frequently leads to better generalization.
5.Validation and Regularization: Methods like cross-validation and regularization are applied to avoid overfitting. Using cross-validation, you may make sure the model works well on various subsets by dividing the data into distinct sets. Excessively complex models are penalized by regularization approaches like L1 and L2 regularization.    

Problem 2.8
```

```python
# Problem 2.8
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
# Step 1: Create a DataFrame manually
data = {
    'Age': [25, 56, 65, 32, 41, 49],
    'Income ($)': [49000, 156000, 99000, 192000, 39000, 57000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("\n")

# Step 2: Normalize the data using scikit-learn
scaler = StandardScaler()
norm_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)

print("Normalized DataFrame:")
print(norm_df)
print("\n")

# Show calculations
print("Calculations:")
print("Mean of Age:", scaler.mean_[0])
print("Standard Deviation of Age:", np.sqrt(scaler.var_[0]))
print("Mean of Income ($):", scaler.mean_[1])
print("Standard Deviation of Income ($):", np.sqrt(scaler.var_[1]))
print("\n")

print("Normalized Age values:")
print(norm_df['Age'])
print("\n")

print("Normalized Income ($) values:")
print(norm_df['Income ($)'])
```
